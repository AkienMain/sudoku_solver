```python
import cv2
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 4, figsize=(16, 40))

img_in = cv2.imread('input/sample.png')
img_dist = cv2.imread('output/sample.png/corrected_distortion.png')
img_sep = cv2.imread('output/sample.png/separated_images.png')
img_out = cv2.imread('output/sample.png/output.png')

axs[0].imshow(img_in)
axs[1].imshow(img_dist)
axs[2].imshow(img_sep)
axs[3].imshow(img_out)

axs[0].set_title('Input')
axs[1].set_title('Corrected Distortion')
axs[2].set_title('Separated Images')
axs[3].set_title('Output')
```
